```python
# for plotting
import matplotlib.pyplot as plt
from matplotlib.image import imread
from PIL import Image
from PIL import ImageDraw
import numpy as np
import os

%matplotlib inline
plt.rcParams['figure.figsize'] = (3, 4)

# for drawing box - https://wikidocs.net/73720
#   img : Image.open(*)으로 load한 이미지
#   c : list of int
def pil_draw_rect(img, c):
    draw = ImageDraw.Draw(img)
    draw.rectangle(((c[0], c[1]), (c[2], c[3])), outline=(100, 255, 255), width=4)
    return img

# for calculating bounding box coordinate
#   line : single string
def bx_coord(line, height=480, width=640):
    # w, h = 640, 480
    n_s = line.split()
    n = [float(i) for i in n_s]
    box_f = (width*(n[1]-n[3]/2), height*(n[2]-n[4]/2), width*(n[1]+n[3]/2), height*(n[2]+n[4]/2))
    box = [int(i) for i in box_f]
    return box
```

```python
import numpy as np
    
def get_path(idx):
    dir_path = "/home/ai_competition6/name/" + str(idx) + ".txt"
    with open(dir_path, "r") as f:
        paths = f.readlines()
        num = len(paths)
        index = np.random.randint(num)
    if paths[index][-1] == '\n':
        return paths[index][:-1]
    else:
        return paths[index]
```

```python
import PIL
from PIL import Image


def merge_simple(images_idx, color = False):
    
    make_size = (640, 480)
    new_image = Image.new('RGB', make_size, (255,255,255))
    labels = []
    
    for i in range(4):
        path = get_path(images_idx[i])
        img_path = path + '.jpg'
        txt_path = path + '.txt'
        img = Image.open(img_path)
        with open(txt_path, "r") as f:
            line = f.readlines()
            # caption 맨뒤에 줄바꿈 있으면 무시
            if line[0][-1] == '\n':
                line = line[0][:-1]
            else:
                line = line[0]
        t = line.split(' ')
        name, center_x, center_y, img_w, img_h = tuple(map(float, t))
        w, h = img.size
        if img_w > 0.5 or img_h > 0.5:
            resized_img = img.resize((320, 240), Image.ANTIALIAS)
            new_image.paste(resized_img,(int(make_size[0]*(0.5*(i//2))),int(make_size[1]*(0.5*(i%2)))))
            label = [int(name), center_x/2 + 0.5*(i//2), center_y/2 + 0.5*(i%2), img_w/2, img_h/2]
            label = map(str, label)
            labels.append(' '.join(label))
        else:
            x = np.random.uniform(center_x + img_w/2 - 0.5, center_x - img_w/2)
            y = np.random.uniform(center_y + img_h/2 - 0.5, center_y - img_h/2)
            b = ' '.join(map(str,[name,x + 0.25, y + 0.25, 0.5, 0.5]))
            box = bx_coord(b, height=480, width=640)
            cropped = img.crop(box)
            new_image.paste(cropped,(int(make_size[0]*(0.5*(i//2))),int(make_size[1]*(0.5*(i%2)))))
            label = [int(name), center_x - x + 0.5*(i//2), center_y - y + 0.5*(i%2), img_w, img_h]
            label = map(str, label)
            labels.append(' '.join(label))
        
        new_label = '\n'.join(labels)
    
    # color filter
    if color == True:
        pixels = new_image.load()
        new_w, new_h = new_image.size
        rs = np.random.uniform(8,13)
        gs = np.random.uniform(8,13)
        bs = np.random.uniform(8,13)
        for px in range(new_w):
            for py in range(new_h):
                r, g, b = new_image.getpixel((px, py))
                pixels[px, py] = (int(r*rs/10), int(g*gs/10), int(b*bs/10))
                
    return new_image, new_label


from random import sample

idx = sample(range(60), 4)
nimg, ntxt = merge_simple(idx, color = True)
plt.figure(figsize=(16, 18))

labels = ntxt.split('\n')
for l in labels:        
    nimg = pil_draw_rect(nimg, bx_coord(l, height=480, width=640))
plt.imshow(nimg)
```

## 고찰
test set에는 초록 노랑 빨강 배경의 사진들이 있다

색깔이 있는 것은 640x480해상도이고

색깔이 없는 것은 1920x1080해상도이다

cfg파일을 보면 width=416 height=416으로 되어있는데 network가 알아서 (416,416)으로 resize해주는 것

한 물체는 최대 하나씩만 나옴

### 1. input 파일
    - backsub 이미지 (640x480, .jpg)
    - backsub 이미지의 box 중심,길이 데이터 (.txt)
    
### 2. 구현방식
    - 4~5개의 틀을 만들고
    - 해당 틀에 들어가는 n개의 이미지를 선택하여 조합 생성
    - ex. 위2개/아래3개 모양의 틀을 1번, 위2개/중간2개/아래2개 모양의 틀을 2번이라 할 경우,
      1번틀에 들어가는 이미지 5개를 선택해 이미지 1000개 생성,
      2번틀에 들어가는 이미지 6개를 선택해 이미지 1000개 생성
      이런식으로 이미지 조합 여러개 생성 가능
      
### 3. output 파일
    - 640x480의 새로운 조합 이미지 (.jpg)
    - 해당 이미지의 box 데이터 (.txt)

```python
import PIL
from PIL import Image

def img_txt_path(dir_path, idx):
    backup_filelist = sorted(os.listdir(dir_path))
    directory = backup_filelist[idx]
    backup_file_path = dir_path + '/' + directory + '/'
    backup_file_list = sorted(os.listdir(backup_file_path))
    num = len(backup_file_list) // 2
    # 각 dir중에 sementation 이미지 중 하나 선택
    num_idx = np.random.randint(num)
    img_path = backup_file_path + backup_file_list[num_idx*2]
    txt_path = backup_file_path + backup_file_list[num_idx*2 + 1]
    # print(img_path)
    # print(txt_path)
    return img_path, txt_path


def merge_image(dir="/home/ai_competition6/3.backsub_images_100/", merge_num = 10, noise = False, crop = False, ratio = False):
    '''
    중복없이 merge_num가지의 물체를 merge함
    input: dir and number of merge
    
    output: merge된 new image와 label txt
    
    '''
    a = set()
    if ratio == False:
        make_size = (640, 480)
    else:
        make_size = (640, 420)
    # create new empty image
    b1 = np.random.randint(50)
    b2 = np.random.randint(50)
    b3 = np.random.randint(50)
    new_image = Image.new('RGB', make_size, (10+5*b1,10+5*b2,10+5*b3))
    labels = []
    filled = []
    
    
    while len(a) < merge_num:
        idx = np.random.randint(60)
        if idx in a:
            continue
        else:
            a.add(idx)
        img_path, txt_path = img_txt_path(dir, idx)
        img = Image.open(img_path)
        img = img.convert("RGBA")
        datas = img.getdata()
        newData = []
        cutOff = 7
        for item in datas:
            if item[0] <= cutOff and item[1] <= cutOff and item[2] <= cutOff:
                newData.append((255, 255, 255, 0))
                # RGB의 각 요소가 모두 cutOff 이상이면 transparent하게 바꿔줍니다
            else:
                newData.append(item)
                # 나머지 요소는 변경하지 않습니다.
        img.putdata(newData)
        f = open(txt_path, "r")
        line = f.readlines()
        # caption 맨뒤에 줄바꿈 있으면 무시
        if line[0][-1] == '\n':
            line = line[0][:-1]
        else:
            line = line[0]
        f.close()
        t = line.split(' ')
        name, center_x, center_y, img_w, img_h = tuple(map(float, t))
        w, h = img.size
        
        # cropped 이미지의 크기 조절
        scale = np.random.uniform(0.4, 0.6)
        
        size = (int(w*img_w*scale) , int(h*img_h*scale))
        cnt = 0
        while True:
            cnt += 1
            if cnt == 100:
                break
            x = np.random.uniform(img_w*scale/2, 1-img_w*scale/2)
            y = np.random.uniform(img_h*scale/2, 1-img_h*scale/2)
            # 이미 칠해진 box안에 중심점이 들어가지 않도록 방지함
            again = False
            for box in filled:
                # 만약 중심점이 이미 칠해진 box안에 있는 경우
                if box[1] - box[3]/2 <= x <= box[1] + box[3]/2 and box[2] - box[4]/2 <= y <= box[2] + box[4]/2:
                    again = True
                # 만약 이미 칠해진 box들의 중심점이 이번 이미지 안에 들어갈 경우
                if x - size[0]/w/2 <= box[1] <= x + size[0]/w/2 and y - size[1]/h/2 <= box[2] <= y + size[1]/h/2:
                    again = True
            
            if again == False:
                break
        if cnt == 100:
            break
        # cut image along the anchor box
        box = bx_coord(line, height=h, width=w)
        cropped = img.crop(box)
        if crop == True:
            p = cropped.load()
            c_w, c_h = cropped.size
            div_x = np.random.randint(5)
            div_y = np.random.randint(5)
            for px in range(c_w*div_x//5, c_w*(div_x+1)//5):
                for py in range(c_h*div_y//5, c_h*(div_y+1)//5):
                    p[px, py] = (10+5*b1,10+5*b2,10+5*b3)
        resized_cropped = cropped.resize(size, Image.ANTIALIAS)
        cropped.thumbnail(size, Image.ANTIALIAS)
        #paste cropped image to empty image
        new_image.paste(cropped,(int(make_size[0]*(x-size[0]/make_size[0]/2)),int(make_size[1]*(y-size[1]/make_size[1]/2))), mask=cropped)
        label = [int(name), x, y, size[0]/make_size[0], size[1]/make_size[1]]
        filled.append(label)
        label = map(str, label)
        labels.append(' '.join(label))
        new_label = '\n'.join(labels)

    # 필터 씌우기
    if noise == True:
        pixels = new_image.load()
        new_w, new_h = new_image.size
        for px in range(new_w):
            for py in range(new_h):
                r, g, b = new_image.getpixel((px, py))
                rs = np.random.uniform(8,13)
                gs = np.random.uniform(8,13)
                bs = np.random.uniform(8,13)
                pixels[px, py] = (int(r*rs/10), int(g*gs/10), int(b*bs/10))
    else:
        pixels = new_image.load()
        new_w, new_h = new_image.size
        rs = np.random.uniform(5,15)
        gs = np.random.uniform(5,15)
        bs = np.random.uniform(5,15)
        for px in range(new_w):
            for py in range(new_h):
                r, g, b = new_image.getpixel((px, py))
                pixels[px, py] = (int(r*rs/10), int(g*gs/10), int(b*bs/10))
        
    return new_image, new_label, ratio
```
